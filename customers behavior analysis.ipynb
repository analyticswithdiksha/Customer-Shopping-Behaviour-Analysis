```python
from sqlalchemy import create_engine
import pandas as pd

# Step 1: Connect to MySQL - database ab exist karta hai
username = "root"           
password = "root123" # <- APNA PASSWORD DAALO
host = "localhost"          
port = "3306"               
database = "customer_db"    

engine = create_engine(f"mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}")

# Step 2: CSV load karo aur upload kar do
df = pd.read_csv(r'C:\Users\Welcome\Desktop\customer2.csv', encoding='utf-8-sig') # <- YAHAN PURA KAR
df.columns = ['id', 'age', 'gender', 'income', 'education', 'region', 
              'loyalty_status', 'purchase_frequency', 'purchase_amount', 
              'product_category', 'promotion', 'satisfaction_score']

table_name = "customers"
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
